```python
concat_frame=[new_data_1,new_data_2]
new_data=pd.concat(concat_frame).dropna(axis=0)
```

```python
new_data.head(10)
```

```python
corr_data=new_data.copy()
corr_data=corr_data.corr()#->corrlation data
```

```python
labels=[list(new_data['state_abbreviation'].unique())]
labels=labels[0]
```

```python
fig=plt.figure(figsize=(12,12))
ax=fig.add_subplot(111)

ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(labels)))

ax.set_yticklabels(labels)
ax.set_xticklabels(labels)
ax.matshow(corr_data,cmap=plt.cm.RdYlGn)#<-heap map or heat map "may be not sure"

plt.show()
```
